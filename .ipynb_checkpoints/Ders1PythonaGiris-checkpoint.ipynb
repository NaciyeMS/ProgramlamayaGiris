### üí° Tanƒ±m:

**Veri (Data)**, bir programƒ±n **i≈üleyip anlamlandƒ±rdƒ±ƒüƒ± bilgi par√ßacƒ±klarƒ±dƒ±r**.  
Python'da veriler, **deƒüi≈ükenler (variables)** i√ßinde saklanƒ±r.  
Her veri bir **t√ºre (data type)** sahiptir ‚Äî bu t√ºr, bilgisayara **o bilginin ne anlama geldiƒüini** s√∂yler.

üéØ KISACA
> üí¨ **Veri**, bilgisayarƒ±n i≈ülediƒüi her t√ºrl√º bilgidir.  
>
> üîç **Veri t√ºr√º**, bu bilginin ne olduƒüunu (metin mi, sayƒ± mƒ±, mantƒ±ksal mƒ±) belirler.  
>
> üß© **Deƒüi≈üken**, bu verileri saklamak i√ßin kullanƒ±lan **‚Äúetiketli kutu‚Äù** gibidir.


### üì¶ Deƒüi≈ükenlerin ƒ∞√ßinde Saklanan Bilgi T√ºrleri

| Veri T√ºr√º | A√ßƒ±klama | √ñrnekler |
|:----------:|:---------|:----------|
| üìù **Metinler (String)** | Harf, kelime veya c√ºmlelerden olu≈üur. | `"naciye"`, `"nasƒ±lsƒ±n"` |
| üî¢ **Tam Sayƒ±lar (Integer)** | Ondalƒ±k kƒ±smƒ± olmayan sayƒ±lar. | `1`, `78`, `-25` |
| ‚öôÔ∏è **Ondalƒ±klƒ± Sayƒ±lar (Float)** | Nokta (.) ile ayrƒ±lan ondalƒ±klƒ± sayƒ±lar. | `0.45`, `2.65`, `-1.5` |
| üîò **Mantƒ±ksal (Boolean)** | Doƒüru veya yanlƒ±≈ü deƒüerlerini temsil eder. | `True`, `False` |

### üí° Deƒüi≈üken Nedir?
**Deƒüi≈üken (variable)**, verileri bellekte tutmak i√ßin kullandƒ±ƒüƒ±mƒ±z bir isimdir.  
Bir deƒüi≈üken, bilgisayarda bir bilgiyi ‚Äúetiketli kutu‚Äù gibi saklar.

### üß© Tanƒ±mlama Kuralƒ±

```python

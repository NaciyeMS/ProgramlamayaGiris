### Harf Notu AralÄ±klarÄ±
- 90 â€“ 100 : AA  
- 80 â€“ 89  : BA  
- 70 â€“ 79  : BB  
- 60 â€“ 69  : CB  
- 50 â€“ 59  : CC  
- 0  â€“ 49  : FF  

Fonksiyon, parametre olarak **ortalama** deÄŸerini almalÄ± ve
ortalamanÄ±n karÅŸÄ±lÄ±k geldiÄŸi **harf notunu ekrana yazdÄ±rmalÄ±dÄ±r**.


```python
def harf_notu(ortalama):
    if 90 <= ortalama <= 100:
        print("Harf Notu: AA")
    elif 80 <= ortalama < 90:
        print("Harf Notu: BA")
    elif 70 <= ortalama < 80:
        print("Harf Notu: BB")
    elif 60 <= ortalama < 70:
        print("Harf Notu: CB")
    elif 50 <= ortalama < 60:
        print("Harf Notu: CC")
    elif 0 <= ortalama < 50:
        print("Harf Notu: FF")
    else:
        print("GeÃ§ersiz ortalama!")
    #return kullanÄ±lmazsa fonksiyon iÃ§erisinde yapÄ±lan iÅŸlem oarada sonuÃ§lanmÄ±ÅŸ 
    """gerek veri tabanÄ±na insert edilmiÅŸ olabilir, ekrana yazdÄ±rÄ±lmÄ±ÅŸ olabilir.burada elde edilen verilerin geriye dÃ¶ndÃ¼rÃ¼lmesine gerek kalmamÄ±ÅŸtÄ±r
    """

harf_notu(55)
harf_notu(85)
```

# Pythonâ€™da Yerel (Local) ve Global DeÄŸiÅŸkenler

Python programlarÄ±nda deÄŸiÅŸkenlerin **nerede tanÄ±mlandÄ±ÄŸÄ±**,  
nerede **eriÅŸilebildiÄŸini** belirler.  
Bu kavrama **deÄŸiÅŸken kapsamÄ± (scope)** denir.

---

## 1. Yerel (Local) DeÄŸiÅŸken

Bir fonksiyonun **iÃ§inde tanÄ±mlanan** deÄŸiÅŸkenlere yerel deÄŸiÅŸken denir.

### Ã–zellikleri
- Sadece tanÄ±mlandÄ±ÄŸÄ± **fonksiyon iÃ§inde** geÃ§erlidir
- Fonksiyon dÄ±ÅŸÄ±ndan **eriÅŸilemez**
- Fonksiyon Ã§alÄ±ÅŸmasÄ± bitince **bellekten silinir**

### Ã–rnek


```python
def fonksiyon():
    x = 10   # yerel deÄŸiÅŸken
    print(x)

fonksiyon()
 print(x)  # HATA! x fonksiyon dÄ±ÅŸÄ±nda tanÄ±mlÄ± deÄŸil
```

## 2. Global DeÄŸiÅŸken

FonksiyonlarÄ±n **dÄ±ÅŸÄ±nda tanÄ±mlanan** deÄŸiÅŸkenlere global deÄŸiÅŸken denir.

### Ã–zellikleri
- ProgramÄ±n her yerinden eriÅŸilebilir
- Fonksiyon iÃ§inde **okunabilir**
- Fonksiyon iÃ§inde **deÄŸiÅŸtirilemez** (Ã¶zel durum hariÃ§)


```python
x = 5  # global deÄŸiÅŸken

def fonksiyon():
    print(x)

fonksiyon()
print(x)
```

## 3. Global DeÄŸiÅŸkeni Fonksiyon Ä°Ã§inde DeÄŸiÅŸtirme

Global bir deÄŸiÅŸkeni fonksiyon iÃ§inde deÄŸiÅŸtirmek iÃ§in  
**global anahtar kelimesi** kullanÄ±lÄ±r.


```python
x = 10

def degistir():
    global x
    x = 20

degistir()
print(x)
```

## 4. AynÄ± Ä°simli Yerel ve Global DeÄŸiÅŸkenler

Yerel deÄŸiÅŸken, global deÄŸiÅŸken ile **aynÄ± isimde olsa bile**
Ã¶ncelik **yerel deÄŸiÅŸkendedir**.


```python
x = 100  # global

def fonksiyon():
    x = 50   # yerel
    print("Fonksiyon iÃ§i x:", x)

fonksiyon()
print("Global x:", x)
```

## 5. HatalÄ± KullanÄ±m Ã–rneÄŸi

Global deÄŸiÅŸkeni `global` demeden deÄŸiÅŸtirmeye Ã§alÄ±ÅŸmak hata verir.


```python
x = 10

def fonksiyon():
    x = x + 5  # HATA!
    pass
```

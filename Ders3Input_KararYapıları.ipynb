# Ä°Ã§ Ä°Ã§e If-Else YapÄ±sÄ± (Nested If-Else) â€“ AÃ§Ä±klama ve Ã–rnek

Ä°Ã§ iÃ§e if-else, bir koÅŸulun saÄŸlanmasÄ± durumunda baÅŸka bir koÅŸulu kontrol etmek iÃ§in kullanÄ±lÄ±r.  
Bu sayede daha detaylÄ± kontrol mekanizmalarÄ± oluÅŸturabiliriz.

## Ã–rnek MantÄ±k

- EÄŸer `sayi > 0` ise: sayÄ± pozitiftir  
  - EÄŸer `sayi % 2 == 0` ise: sayÄ± Ã§ifttir  
  - DeÄŸilse: sayÄ± tektir  
- DeÄŸilse, eÄŸer `sayi < 0` ise: sayÄ± negatiftir  
- DeÄŸilse: sayÄ± sÄ±fÄ±rdÄ±r


````python
# KullanÄ±cÄ±dan bir sayÄ± alma
sayi = int(input("Bir sayÄ± girin: "))

# SayÄ±nÄ±n pozitif, negatif veya sÄ±fÄ±r olup olmadÄ±ÄŸÄ±nÄ± kontrol etme
if sayi > 0:
    print("SayÄ± pozitif.")
    # Pozitif sayÄ± ise Ã§ift veya tek olduÄŸunu kontrol et
    if sayi % 2 == 0:
        print("AyrÄ±ca sayÄ± Ã§ifttir.")
    else:
        print("AyrÄ±ca sayÄ± tektir.")
elif sayi < 0:
    print("SayÄ± negatif.")
else:
    print("SayÄ± sÄ±fÄ±rdÄ±r.")
````

# SORU : â“ KullanÄ±cÄ±nÄ±n GirdiÄŸi SayÄ±larÄ± Toplayan Program (Listesiz)

Bu programda kullanÄ±cÄ±ya **kaÃ§ tane sayÄ± gireceÄŸi** sorulacak.  
Daha sonra kullanÄ±cÄ± bu sayÄ±larÄ± teker teker girecek ve biz **liste kullanmadan**, yalnÄ±zca bir toplama deÄŸiÅŸkeni ile tÃ¼m sayÄ±larÄ± toplayacaÄŸÄ±z.

---

## ðŸ“Œ Algoritma

1. **BaÅŸla**  
2. KullanÄ±cÄ±ya kaÃ§ sayÄ± gireceÄŸi sor  
3. `toplam = 0` yap  
4. `i = 1` olarak baÅŸlat  
5. `i <= adet` olduÄŸu sÃ¼rece while dÃ¶ngÃ¼sÃ¼ne gir  
6. KullanÄ±cÄ±dan bir sayÄ± al  
7. Bu sayÄ±yÄ± `toplam` deÄŸiÅŸkenine ekle

# SORU : ðŸ—³ï¸ SeÃ§im Anketi Oy SayÄ±mÄ± ProgramÄ± (Listsiz Ã‡Ã¶zÃ¼m)

Bir seÃ§im anketi **N tane seÃ§mene** uygulanmÄ±ÅŸtÄ±r.  
Her seÃ§men oyunu sayÄ± ile belirtmiÅŸtir:

- **1 â€“ 8** â†’ Partilerin numarasÄ±  
- **9** â†’ KararsÄ±z  
- **0** â†’ HiÃ§biri  

GÃ¶rev:  
Her seÃ§eneÄŸin **aldÄ±ÄŸÄ± oy sayÄ±sÄ±nÄ±** ve **yÃ¼zdesini** hesaplayan bir program yazÄ±nÄ±z.  
Bu soru *liste kullanmadan*, yalnÄ±zca sayaÃ§ deÄŸiÅŸkenleri ile Ã§Ã¶zÃ¼lecektir.

---

## ðŸ“Œ Algoritma

1. **BaÅŸla**  
2. KullanÄ±cÄ±ya toplam seÃ§men sayÄ±sÄ± **N** sor  
3. 0â€“9 arasÄ±ndaki tÃ¼m seÃ§enekler iÃ§in sayaÃ§ deÄŸiÅŸkenlerini **0** yap  
   - `sifir = 0`  
   - `bir = 0`  
   - `iki = 0`  
   - `uc = 0`  
   - `dort = 0`  
   - `bes = 0`  
   - `alti = 0`  
   - `yedi = 0`  
   - `sekiz = 0`  
   - `dokuz = 0`  
4. `i` deÄŸiÅŸkenini **1** yap  
5. `i <= N` olduÄŸu sÃ¼rece dÃ¶ngÃ¼yÃ¼ baÅŸlat  
6. KullanÄ±cÄ±dan 0â€“9 arasÄ± oy al  
7. Girilen sayÄ±ya gÃ¶re ilgili sayacÄ± **1 artÄ±r**  
8. `i` deÄŸerini **1 artÄ±r**  
9. DÃ¶ngÃ¼ bittiÄŸinde her seÃ§eneÄŸin yÃ¼zdesini hesapla:  
   - yÃ¼zde = (sayaÃ§ / N) * 100  
10. SonuÃ§larÄ± ekrana yaz  
11. **Bitir**  

---

## ðŸ’» Python Kod Ã–rneÄŸi 


````python
N = int(input("KaÃ§ seÃ§mene anket uygulandÄ±? "))

# Oy sayaÃ§larÄ±
sifir = bir = iki = uc = dort = bes = alti = yedi = sekiz = dokuz = 0

i = 1
while i <= N:
    oy = int(input(f"{i}. seÃ§menin oyunu girin (0-9): ")) #pusula yaz

    if oy == 0:
        sifir += 1
    elif oy == 1:
        bir += 1
    elif oy == 2:
        iki += 1
    elif oy == 3:
        uc += 1
    elif oy == 4:
        dort += 1
    elif oy == 5:
        bes += 1
    elif oy == 6:
        alti += 1
    elif oy == 7:
        yedi += 1
    elif oy == 8:
        sekiz += 1
    elif oy == 9:
        dokuz += 1

    i += 1

# SonuÃ§larÄ± yazdÄ±rma
print("\n=== OY SONUÃ‡LARI ===")
print("0 (HiÃ§biri):", sifir, "oy   YÃ¼zde:", (sifir / N) * 100)
print("1:", bir, "oy   YÃ¼zde:", (bir / N) * 100)
print("2:", iki, "oy   YÃ¼zde:", (iki / N) * 100)
print("3:", uc, "oy    YÃ¼zde:", (uc / N) * 100)
print("4:", dort, "oy   YÃ¼zde:", (dort / N) * 100)
print("5:", bes, "oy    YÃ¼zde:", (bes / N) * 100)
print("6:", alti, "oy   YÃ¼zde:", (alti / N) * 100)
print("7:", yedi, "oy   YÃ¼zde:", (yedi / N) * 100)
print("8:", sekiz, "oy  YÃ¼zde:", (sekiz / N) * 100)
print("9 (KararsÄ±z):", dokuz, "oy   YÃ¼zde:", (dokuz / N) * 100)
````

# SORU : ðŸ¦  Bakteri Ã‡oÄŸalma Problemi

Bir bakteri tÃ¼rÃ¼ **her 4 dakikada bir ikiye bÃ¶lÃ¼nerek** Ã§oÄŸalmaktadÄ±r.  
BaÅŸlangÄ±Ã§ta **1 adet bakteri** olduÄŸunu kabul ederek, kullanÄ±cÄ±dan alÄ±nan **t saat** sonunda kaÃ§ bakteri oluÅŸacaÄŸÄ±nÄ± hesaplayan programÄ± yazÄ±nÄ±z.

---

## ðŸ“Œ Problemin MantÄ±ÄŸÄ±

- 1 saat = 60 dakika  
- Her 4 dakikada bir bÃ¶lÃ¼nme â†’ yani **her 4 dakikada bir bakteri sayÄ±sÄ± 2 katÄ±na Ã§Ä±kar**  
- Toplam bÃ¶lÃ¼nme sayÄ±sÄ±:  
  \[
  \text{bolunme\_sayisi} = \frac{t \times 60}{4}
  \]
- Son bakteri sayÄ±sÄ±:  
  \[
  \text{bakteri} = 2^{\text{bolunme\_sayisi}}
  \]

---

## ðŸ“Œ Algoritma

1. BaÅŸla  
2. KullanÄ±cÄ±dan saat cinsinden **t** deÄŸerini al  
3. `bolunme_sayisi = (t * 60) / 4` hesapla  
4. `bakteri = 2 ** bolunme_sayisi` hesapla  
5. Sonucu ekrana yaz  
6. Bitir  

---

## ðŸ’» Python Kod Ã–rneÄŸi

````python
t = int(input("KaÃ§ saat sonra bakteri sayÄ±sÄ±nÄ± Ã¶ÄŸrenmek istiyorsunuz? "))

bolunme_sayisi = (t * 60) / 4
bakteri = 2 ** bolunme_sayisi

print("Toplam bakteri sayÄ±sÄ±:", int(bakteri))
````

# DÃ–NGÃœ Ä°LE Ã‡Ã–ZEBÄ°LÄ°R MÄ°YÄ°Z? 
## ðŸ“Œ MantÄ±k

- 1 saat = 60 dakika  
- Her 4 dakikada bir bÃ¶lÃ¼nme olur  
- Toplam bÃ¶lÃ¼nme sayÄ±sÄ±:  
  \[
  \text{bolunme\_sayisi} = \frac{t \times 60}{4}
  \]
- For dÃ¶ngÃ¼sÃ¼nde her bÃ¶lÃ¼nmede bakteri sayÄ±sÄ± 2 katÄ±na Ã§Ä±kacak ÅŸekilde artÄ±rÄ±lÄ±r.

---

## ðŸ“Œ Algoritma

1. **BaÅŸla**  
2. KullanÄ±cÄ±dan **t** (saat) deÄŸerini al  
3. `bolunme_sayisi = (t * 60) / 4` formÃ¼lÃ¼yle toplam bÃ¶lÃ¼nme sayÄ±sÄ±nÄ± hesapla  
4. `bakteri` deÄŸiÅŸkenini **1** yap  
5. `i` deÄŸiÅŸkenini **0** yap  
6. `i < bolunme_sayisi` olduÄŸu sÃ¼rece dÃ¶ngÃ¼yÃ¼ baÅŸlat  
7. Her dÃ¶ngÃ¼de `bakteri` deÄŸerini **2 ile Ã§arp**  
8. `i` deÄŸerini **1 artÄ±r**  
9. DÃ¶ngÃ¼ bitince bakteri sayÄ±sÄ±nÄ± ekrana yaz  
10. **Bitir** 

---

## ðŸ’» Python Kod Ã–rneÄŸi 

````python
t = int(input("KaÃ§ saat sonra bakteri sayÄ±sÄ±nÄ± Ã¶ÄŸrenmek istiyorsunuz? "))

bolunme_sayisi = int((t * 60) / 4)

bakteri = 1
i = 0

while i < bolunme_sayisi:
    bakteri *= 2
    i += 1

print("Toplam bakteri sayÄ±sÄ±:", bakteri)
````

### ðŸ” `for` DÃ¶ngÃ¼sÃ¼ AÃ§Ä±klamasÄ±

`for` dÃ¶ngÃ¼sÃ¼, `while` dÃ¶ngÃ¼sÃ¼nÃ¼n yaptÄ±ÄŸÄ± **tÃ¼m iÅŸleri yapabilir**.  
BaÅŸlangÄ±Ã§, bitiÅŸ (koÅŸul) ve artÄ±ÅŸ iÅŸlemleri aÃ§Ä±sÄ±ndan **iÅŸlevsel olarak fark yoktur**.

AralarÄ±ndaki temel fark **yazÄ±m ÅŸeklidir**:

- `while` dÃ¶ngÃ¼sÃ¼nde  
  - baÅŸlangÄ±Ã§ deÄŸeri ayrÄ± yazÄ±lÄ±r  
  - koÅŸul ayrÄ± yazÄ±lÄ±r  
  - artÄ±ÅŸ iÅŸlemi ayrÄ± yazÄ±lÄ±r  

- `for` dÃ¶ngÃ¼sÃ¼nde ise  
  - baÅŸlangÄ±Ã§  
  - bitiÅŸ  
  - artÄ±ÅŸ  

**tek satÄ±rda (`range`) tanÄ±mlanÄ±r.**

Bu nedenle `for` dÃ¶ngÃ¼sÃ¼:
- Daha **kÄ±sa**
- Daha **okunabilir**
- Hata yapma ihtimali daha **dÃ¼ÅŸÃ¼k** bir yapÄ± sunar.

> Ã–zetle:  
> **`for`, `while`â€™Ä±n yaptÄ±ÄŸÄ± her ÅŸeyi yapar; sadece daha dÃ¼zenli ve kontrollÃ¼ bir yazÄ±m sunar.**


````python
i = 1              # baÅŸlangÄ±Ã§
while i <= 10:     # bitiÅŸ (koÅŸul)
    print(i)       # tekrar eden iÅŸ
    i += 1         # artÄ±ÅŸ
````

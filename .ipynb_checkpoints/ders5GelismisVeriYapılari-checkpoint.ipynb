  KÃ¼me elemanlarÄ±nÄ±n belirli bir sÄ±rasÄ± yoktur.

- **TekrarsÄ±zdÄ±r (Unique)**  
  AynÄ± eleman kÃ¼me iÃ§inde sadece **bir kez** bulunabilir.

- **DeÄŸiÅŸtirilebilir (Mutable)**  
  KÃ¼me elemanlarÄ± eklenebilir veya silinebilir.

- **Ä°ndeksleme Yoktur**  
  Elemanlara indeks numarasÄ± ile eriÅŸilemez.

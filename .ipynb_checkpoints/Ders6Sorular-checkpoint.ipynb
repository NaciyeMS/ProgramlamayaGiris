## Python Programlama Sorusu â€“ Dizi ve Ortalama

KullanÄ±cÄ±dan **n adet sayÄ±** girmesi istenecektir.

### GÃ¶rev

1. KullanÄ±cÄ±nÄ±n girdiÄŸi n sayÄ±yÄ± bir **dizi (liste)** iÃ§erisinde saklayÄ±nÄ±z.
2. Bu sayÄ±larÄ±n **aritmetik ortalamasÄ±nÄ±** hesaplayÄ±nÄ±z.
3. Hesaplanan ortalamayÄ± ekrana yazdÄ±rÄ±nÄ±z.

### Kurallar

- SayÄ±lar kullanÄ±cÄ±dan `input()` ile alÄ±nacaktÄ±r.
- SayÄ±lar bir **liste** iÃ§inde tutulacaktÄ±r.
- Ortalama hesabÄ± yapÄ±lÄ±rken liste elemanlarÄ± kullanÄ±lacaktÄ±r.
- Kod okunabilir ve anlaÅŸÄ±lÄ±r olmalÄ±dÄ±r.


````python
# KullanÄ±cÄ±dan kaÃ§ adet sayÄ± gireceÄŸini al
n = int(input("KaÃ§ adet sayÄ± gireceksiniz? "))

sayilar = []  # SayÄ±larÄ± saklayacaÄŸÄ±mÄ±z liste

# KullanÄ±cÄ±dan n adet sayÄ± al
for i in range(n):
    sayi = float(input(f"{i+1}. sayÄ±yÄ± giriniz: "))
    sayilar.append(sayi)

# Ortalama hesaplama
toplam = sum(sayilar)
ortalama = toplam / n

# Sonucu ekrana yazdÄ±r
print("Girilen sayÄ±lar:", sayilar)
print("Aritmetik Ortalama:", ortalama)
````

## Python Programlama Sorusu â€“ SeÃ§im Anketi Analizi

Bir seÃ§im anketi **N** tane seÃ§mene uygulanmÄ±ÅŸtÄ±r.  
SeÃ§menler oylarÄ±nÄ± aÅŸaÄŸÄ±daki ÅŸekilde belirtmiÅŸlerdir:

- **1 â€“ 8** : Ä°lgili partinin numarasÄ±  
- **9** : KararsÄ±z  
- **0** : HiÃ§biri  

SeÃ§menlerin verdiÄŸi oylar bir liste iÃ§erisinde tutulmaktadÄ±r.

### GÃ¶rev

Python kullanarak aÅŸaÄŸÄ±daki iÅŸlemleri gerÃ§ekleÅŸtiren bir program yazÄ±nÄ±z:

1. SeÃ§men oylarÄ±nÄ± **liste** olarak alÄ±nÄ±z.
2. Her bir seÃ§eneÄŸin:
   - AldÄ±ÄŸÄ± **oy sayÄ±sÄ±nÄ±**
   - Toplam oy iÃ§indeki **yÃ¼zdesini**
   hesaplayÄ±nÄ±z.
3. SonuÃ§larÄ± **anlaÅŸÄ±lÄ±r ve dÃ¼zenli** bir biÃ§imde ekrana yazdÄ±rÄ±nÄ±z.

### Kurallar

- Dizi (liste) zorunlu olarak kullanÄ±lacaktÄ±r.
- Oy sayÄ±mÄ± iÃ§in **if-else ile tek tek sayma yapÄ±lmayacaktÄ±r**.
- Pythonâ€™un **hazÄ±r veri yapÄ±larÄ±ndan** yararlanÄ±lmalÄ±dÄ±r.
- Kod, okunabilir ve yorum satÄ±rlarÄ± iÃ§ermelidir.

### Ã–rnek Girdi

```python
oylar = [1, 3, 3, 9, 0, 5, 1, 8, 3, 9, 1, 0, 7]


````python
## Python Ã‡Ã¶zÃ¼m â€“ Dict Kullanarak SeÃ§im Anketi


# SeÃ§men sayÄ±sÄ±nÄ± al
n = int(input("Toplam seÃ§men sayÄ±sÄ±nÄ± giriniz: "))

# OylarÄ± tutacak liste
oylar = []

# KullanÄ±cÄ±dan oylarÄ± al
print("OylarÄ± giriniz (0-9 arasÄ±):")
for i in range(n):
    oy = int(input(f"{i+1}. seÃ§menin oyu: "))
    oylar.append(oy)

# Oy sayÄ±mlarÄ±nÄ± tutacak sÃ¶zlÃ¼k (dict)
sayac = {}

# OylarÄ± say
for oy in oylar:
    if oy in sayac:
        sayac[oy] += 1
    else:
        sayac[oy] = 1

# SonuÃ§larÄ± yazdÄ±r
print("\n--- Anket SonuÃ§larÄ± ---")
for secim, adet in sayac.items():
    yuzde = (adet / n) * 100

    if secim == 9:
        ad = "KararsÄ±z"
    elif secim == 0:
        ad = "HiÃ§biri"
    else:
        ad = f"Parti {secim}"

    print(f"{ad}: {adet} oy (%{yuzde:.2f})")
````

## Python Programlama Sorusu â€“ Ã–ÄŸrenci NotlarÄ± ve Harf Notu

Bir sÄ±nÄ±fta **n Ã¶ÄŸrenci** bulunmaktadÄ±r.  
Her Ã¶ÄŸrenci iÃ§in **adÄ±**, **vize** ve **final** notlarÄ± girilecektir.

### Not Hesaplama KurallarÄ±

- Vize notunun %40â€™Ä±
- Final notunun %60â€™Ä±  
alÄ±narak **ortalama** hesaplanacaktÄ±r.

### Harf Notu AralÄ±klarÄ±

- 90 â€“ 100 : AA  
- 80 â€“ 89  : BA  
- 70 â€“ 79  : BB  
- 60 â€“ 69  : CB  
- 50 â€“ 59  : CC  
- 0  â€“ 49  : FF  

### GÃ¶rev

1. KullanÄ±cÄ±dan **Ã¶ÄŸrenci sayÄ±sÄ±nÄ± (n)** alÄ±nÄ±z.
2. Her Ã¶ÄŸrenci iÃ§in:
   - Ã–ÄŸrenci adÄ±nÄ±
   - Vize notunu
   - Final notunu  
   kullanÄ±cÄ±dan alÄ±nÄ±z.
3. Ã–ÄŸrenci bilgilerini **liste iÃ§inde sÃ¶zlÃ¼k (dict)** yapÄ±sÄ± kullanarak saklayÄ±nÄ±z.
4. Her Ã¶ÄŸrencinin:
   - OrtalamasÄ±nÄ±
   - Harf notunu  
   hesaplayÄ±nÄ±z.
5. Ã–ÄŸrenci adÄ±, ortalamasÄ± ve harf notunu ekrana yazdÄ±rÄ±nÄ±z.

### Kurallar

- Ã–ÄŸrenci bilgileri **liste + dict** ile tutulacaktÄ±r.
- `if-elif` yapÄ±sÄ± harf notu belirlemek iÃ§in kullanÄ±lacaktÄ±r.
- Kod okunabilir ve aÃ§Ä±klamalÄ± olmalÄ±dÄ±r.


````python
ogrenciler = []

n = int(input("Ã–ÄŸrenci sayÄ±sÄ±nÄ± giriniz: "))

for i in range(n):
    print(f"\n{i+1}. Ã¶ÄŸrenci")
    ad = input("Ã–ÄŸrenci adÄ±: ")
    vize = float(input("Vize notu: "))
    final = float(input("Final notu: "))

    ortalama = vize * 0.4 + final * 0.6

    if ortalama >= 90:
        harf = "AA"
    elif ortalama >= 80:
        harf = "BA"
    elif ortalama >= 70:
        harf = "BB"
    elif ortalama >= 60:
        harf = "CB"
    elif ortalama >= 50:
        harf = "CC"
    else:
        harf = "FF"

    ogrenci = {
        "ad": ad,
        "vize": vize,
        "final": final,
        "ortalama": ortalama,
        "harf_notu": harf
    }

    ogrenciler.append(ogrenci)

print("\n--- Ã–ÄŸrenci Not Listesi ---")
for ogr in ogrenciler:
    print(f"{ogr['ad']} -> Ortalama: {ogr['ortalama']:.2f}, Harf Notu: {ogr['harf_notu']}")
````
